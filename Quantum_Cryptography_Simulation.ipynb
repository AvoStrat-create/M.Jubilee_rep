```python
import random
```

```python
n = 20
```

```python
#Alice creates random bits
alice_bits = []

for i in range(n):
    bit = random.randint(0, 1)
    alice_bits.append(bit)

print("Alice Bits:")
print(alice_bits)
```

```python
#Alice chooses random bases
alice_bases = []

for i in range(n):
    basis = random.choice(['+', 'x'])
    alice_bases.append(basis)

print("\nAlice Bases:")
print(alice_bases)
```

```python
# Bob chooses random bases
bob_bases = []

for i in range(n):
    basis = random.choice(['+', 'x'])
    bob_bases.append(basis)

print("\nBob Bases:")
print(bob_bases)
```

```python
#Bob measures the qubits
bob_results = []

for i in range(n):

    # Bases match
    if alice_bases[i] == bob_bases[i]:
        bob_results.append(alice_bits[i])

    # Bases do not match
    else:
        random_bit = random.randint(0, 1)
        bob_results.append(random_bit)

print("\nBob Results:")
print(bob_results)
```

```python
#Find matching bases
matching_positions = []

for i in range(n):
    if alice_bases[i] == bob_bases[i]:
        matching_positions.append(i)

print("\nMatching Positions:")
print(matching_positions)
```

```python
#Generating a shared secret key
shared_key = []

for position in matching_positions:
    shared_key.append(bob_results[position])

print("\nShared Secret Key:")
print(shared_key)
```

```python
#Eve the hacker
eve_bases = []

for i in range(n):
    basis = random.choice(['+', 'x'])
    eve_bases.append(basis)

eve_results = []

for i in range(n):

    # Eve guesses correctly
    if alice_bases[i] == eve_bases[i]:
        eve_results.append(alice_bits[i])

    # Eve guesses incorrectly
    else:
        eve_results.append(random.randint(0, 1))
```

```python
#Bob receives Eve's modified qubits
bob_results = []

for i in range(n):

    if eve_bases[i] == bob_bases[i]:
        bob_results.append(eve_results[i])

    else:
        bob_results.append(random.randint(0, 1))
```

```python
#Detect the hacker
errors = 0
checked_bits = 0

for i in matching_positions:

    checked_bits += 1

    if alice_bits[i] != bob_results[i]:
        errors += 1

print("\nErrors Detected:", errors)

if errors > 0:
    print("WARNING: Eve detected!")
else:
    print("Communication secure.")
```
